```python
f = pd.read_csv('bigrams.csv', encoding='utf-8')
```

```python
import string
invalidChars = set(string.punctuation)
```

```python
freq_filtered = f.loc[(f['frequency'] >= 3)]
```

```python
freq_filtered.head()
```

```python
raw_freq = []
lr = []
ngrams = []
pmi = []
freq = []
t = []
tag = []
```

```python
ngram_list = freq_filtered.ngram.tolist()
freq_list = freq_filtered['raw frequency'].tolist()
lr_list = freq_filtered['likelihood ratio'].tolist()
pmi_list = freq_filtered['pmi'].tolist()
raw_freq_list = freq_filtered['frequency'].tolist()
t_list = freq_filtered['t score'].tolist()
tag_list = freq_filtered['tag'].tolist()
for i in range(len(ngram_list)):
    if not any(char in invalidChars for char in ngram_list[i]): 
        ngrams.append(ngram_list[i])
        freq.append(freq_list[i])
        lr.append(lr_list[i])
        pmi.append(pmi_list[i])
        t.append(t_list[i])
        raw_freq.append(raw_freq_list[i])
        tag.append(tag_list[i])
        
```

```python
len(ngrams) == len(raw_freq) == len(lr) == len(pmi) == len(freq) == len(t) == len(tag)
```

```python
df=pd.DataFrame({'ngram':ngrams, 
                 'tag':tag, 
                 'frequency':freq, 
                 'raw frequency':raw_freq, 
                 'likelihood ratio':lr, 
                 'pmi':pmi, 
                 't score':t})
```

```python
import csv
```

```python
CORPORA = ['open', 'rus']
```

```python
def to_df(filepath, corpus_type):
    with open(filepath, 'r', encoding='utf-8') as fp:
        reader = csv.reader(fp, delimiter=';', quotechar='"')
        data_read = [row for row in reader]
    if corpus_type == 'open':
        ngram = []
        freq = []
        for row in data_read:
            if int(row[1]) >= 3:
                ngram.append(row[0])
                freq.append(int(row[1]))
        df = pd.DataFrame({'ngram':ngram, 'freq':freq})
        return df
    elif corpus_type == 'rus':
        ngram = []
        freq = []
        for row in data_read:
            if '0' in row:
                index = row.index('0')
                gram = ' '.join(row[1:index])
                if not any(char in invalidChars for char in gram):
                    ngram.append(gram)
                    try:
                        freq.append(int(row[0]))
                    except ValueError:
                        freq.append(int(row[0][1:]))
        df = pd.DataFrame({'ngram':ngram, 'freq':freq})
        return df
```

```python
open2_df = to_df('openc_bigrams.csv', CORPORA[0])
```

```python
open3_df = to_df('openc_trigrams.csv', CORPORA[0])
```

```python
rus2_df = to_df('rusc_bigrams.csv', CORPORA[1])
```

```python
rus3_df = to_df('rusc_trigrams.csv', CORPORA[1])
```
